```python
# [Answer to Ex. 5.1.2]
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns 

iris.head(5)
g = sns.regplot(y="sepal_length", x="petal_length", fit_reg=True, order=2, data=iris)
plt.ylabel("Sepal length")
plt.xlabel("Petal length")
plt.title("Petal and Sepal length relation")       
plt.show(g)
#Yet, due to no petal having a length between 2-3, the second order polynomial might not be the best tool.
```
