## Data source
In order to analyse the **[INSERT PART FROM PROBLEMFORMULERING STATING WHAT WE WANT TO ANALYSE]** we scraped historical data of all job postings on JobIndex from 2007-2017, including both years. JobIndex is the largest Danish job portal which works as an index, by fetching job postings from hundreds of different sources. Their archive, the [JobIndex archive](https://www.jobindex.dk/jobsoegning?jobage=archive), contains all the job postings listed on their site since 1998, yet the amount of job postings prior to 2007 is very scarce (contains approximately 13.000 jobs postings), which is why our dataset only contains data from 2007 and onwards. The job postings are also divided into different characteristics: Area, Job Category and Job Type  (where Job Type is a subdivision of Job Category). Based on the archive we are able to obtain the number of job posting across year, area, job category and job type.   
JobIndex has defined 13 different areas and 10 different Job Categories, where each job category contains various job types.

## Web scraping / data colelction
By inspecting the site it self and elements of the  archive's website we were able to attain the part of the html code which contained the amount of job postings. Firstly, specifying the search across each years, area, job category and job type would alter the URL itself, enabling us to create a function which based on the four inputs (year, area, job category and job type) would give as a useable URL as output. Afterwards, a list of areas and a dictionary of job category and corresponding job types was created. Based on the function, the list and the dictionary we were able to create a full list of all the URL's which would be needed in order to fetch the amount of job postings (by looping) across each year, area, job category and job type. Here, it's important to be aware of multiple categories using the same URL. E.g. The Job Category "Retail management" occurs under the job category of both "Trade and service" as well as "Management and staff", yet they share the same URL. Then based on all the URL's e were able to scrape the amount of job postings across year, area, job category and job type by parsing through the HTML code and searching for the `h2` header and the class `search-title`. Finally, the amount of job postings were added to a DataFrame, which contained a column for year, area, job category and job type, whereas the row would be the different combinations thereof. i.e.:

## Data limitations
Despite of it's glory, the JobIndex archive doesn't contain information of which cities the companies of the job postings are from nor which cities the actual jobs will take place in. Only half of the job postings in the JobIndex archive contain any kind of city-related information and as such, narrowing the labor demand (based on job postings) down to cities instead of areas is not entirely possible. Furthermore, due to the job categorization in the JobIndex archive, some of the job postings are ambiguous and will occur in more than one job category. Therefore, when summing all the job postings based on their Job Category (or Job Type) will result in counting duplicates. Therefore it is important to distinguish between the amount of job postings in each category and the total amount of job postings. In order to clarify this, our dataset will contain variables denoted as `"N_[specification]"` where `[specification]` is a general term for whichever characteristics is used (e.g. area).
