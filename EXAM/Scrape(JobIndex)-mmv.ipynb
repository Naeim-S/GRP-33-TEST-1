```python
years = list(range(2007,2018,1))

len_områder = int(len(years*len(links)) / len(område))
len_sektor = int(len(years*len(links)) / sum([len(x) for x in sektor.values()]))
len_jobtype = int(len(years*len(links)) / sum([len(x) for x in sektor.values()]))

områder = [] #Områderne omsættes til én lang liste (se selv)
for x in område:
    områder.append([x]*len_områder)
områder = sum(områder,[])

jobtyper = [] #Jobtyperne fra dictionary sammensættes til én list
for x in sektor.values():
    jobtyper.append(x)
jobtyper=sum(jobtyper, [])

sektorer1 = [] #Sektorerne fra dictionary sammensættes til én list, hvor de fremgår ligeså mange gange som antallet af jobtyper tilknyttet
for key, value in sektor.items():
#     print(key, len([item for item in value if item]))
    sektorer1.append([key]*len([item for item in value if item]))
sektorer=sum(sektorer1, [])

df = pd.DataFrame({'År':years*len(links),
                   'Område':områder,
                   'Sektor':sektorer*len_sektor,
                   'Jobtype':jobtyper*len_jobtype})

df.sort_values(by=['Område', 'År', 'Sektor'])
```
