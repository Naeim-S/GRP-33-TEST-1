## Lo and behold the sacred data archives
In order to analyse the **[INSERT PART FROM PROBLEMFORMULERING STATING WHAT WE WANT TO ANALYSE]** we scraped historical data of all job postings on JobIndex from 2007-2017, including both years. JobIndex is the largest Danish job portal which works as an index, by fetching job postings from hundreds of different sources. Their archive, the [JobIndex archive](https://www.jobindex.dk/jobsoegning?jobage=archive), contains all the job postings listed on their site since 1998, yet the amount of job postings prior to 2007 is very scarce (contains approximately 13.000 jobs postings), which is why our dataset only contains data from 2007 and onwards. The job postings are also divided into different characteristics: Area, Job Category and Job Type  (where Job Type is a subdivision of Job Category).   
JobIndex has defined 13 different areas and 10 different Job Categories, where each category contains varying job types.

## The art of web-scraping

- Hvordan vi scrapede (rent teknisk og ift. kode)
    - We also had to be aware of multiple categories using the same URL. E.g. The Job Category "trade and service" contains the Job Type "Retail management" whereas the Job Category "Management & Staff" also contains the same Job Type "Retail management", yet they share the same URL.   
    As such, URL-duplicates have been removed and the amount of job postings will be placed in the category specified by the URL. E.g. The job postings under job type "Retail management" will only occur in the job category of "Management & Staff" and not under the job category "Commerce & Service" due to its [URL](https://www.jobindex.dk/jobsoegning/ledelse/detailledelse?jobage=archive)

## Data limitations
Despite of it's glory, the JobIndex archive doesn't contain information of which cities the companies of the job postings are from nor which cities the actual jobs will take place in. Only half of the job postings in the JobIndex archive contain any kind of city-related information and as such, narrowing the labor demand (based on job postings) down to cities instead of areas is not entirely possible. Furthermore, due to the job categorization in the JobIndex archive, some of the job postings are ambiguous and will occur in more than one job category. Therefore, when summing all the job postings based on their Job Category (or Job Type) will result in counting duplicates. Therefore it is important to distinguish between the amount of job postings in each category and the total amount of job postings. In order to clarify this, our dataset will contain variables denoted as `"N_[specification]"` where `[specification]` is a general term for whichever characteristics is used (e.g. area).
