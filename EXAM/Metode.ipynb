## Data source
In order to analyse the **[*INSERT PART FROM PROBLEMFORMULERING STATING WHAT WE WANT TO ANALYSE*]** we scraped historical data of all job postings on JobIndex from 2007-2017, including both years. JobIndex is the largest Danish job portal which works as an index, by fetching job postings from hundreds of different sources. Their archive, the [JobIndex archive](https://www.jobindex.dk/jobsoegning?jobage=archive), contains all the job postings listed on their site since 1998, yet the amount of job postings prior to 2007 is very scarce (contains approximately 13.000 jobs postings), which is why our dataset only contains data from 2007 and onwards. The job postings are also divided into different characteristics: Area, Job Category and Job Type  (where Job Type is a subdivision of Job Category). Based on the archive we are able to obtain the number of job posting across year, area, job category and job type.   
JobIndex has defined 13 different areas and 10 different Job Categories, where each job category contains various job types. 
Jobindex as data source allows for analysis of labor demand across time and industries. As society progress demand for labor likely changes as well. The amount of labor demanded within each industry vary. Processing data for demand side of the labor market enables analysis of instant changes in geograchical restructerings and workers required skill levels for instance. Labor supply reacts slowly to changes in demand, which is among causes of structural unemployment. Hence data for labor demand allows for greater insight into the future of the Danish labor market.

## Web scraping / data collection
By inspecting the elements of the  archive's website we were able to attain the part of the html code which contained the amount of job postings. Firstly, specifying the search across each year, area, job category and job type would alter the URL itself, enabling us to create a function returning an URL to a specific job type for each category and area and which takes the four characteristics as inputs (i.e. year, area, job category and job type).  
Here, it was important to be aware of multiple categories using the same URL. E.g. The Job Type "Retail management" occurs under the job category of both "Trade and service" as well as "Management and staff", yet they share the same URL. As such, URL-duplicates have been removed and the amount of job postings will be placed in the category specified by the URL.   

Then, based on all the URL's we were able to scrape the amount of job postings across year, area, job category and job type by parsing through the HTML code and searching for the `h2` header and the class `search-title`. In order to collect such amount of data we made sure to use a delay when scraping Jobindex.dk. Finally, a DataFrame which contained a column for year, area, job category and job type was created, and the amount of job postings across all characteristics were added to the DataFrame as a seperate column. As such, each row would be the different combinations of characteristics. i.e.:

## Data limitations
Despite of it's glory, the JobIndex archive doesn't contain information of which cities the companies of the job postings are from nor which cities the actual jobs will take place in. Only half of the job postings in the JobIndex archive contain any kind of city-related information and as such, narrowing the labor demand (based on job postings) down to cities instead of areas is not entirely possible. Furthermore, due to the job categorization in the JobIndex archive, some of the job postings are ambiguous and will occur in more than one job category. Therefore, when summing all the job postings based on their Job Category (or Job Type) will result in counting duplicates. Therefore it is important to distinguish between the amount of job postings in each category and the total amount of job postings. In order to clarify this, our dataset will contain variables denoted as `"N_[specification]"` where `[specification]` is a general term for whichever characteristics is used (e.g. number of jobpostings in that area for that year or within that job type, area and year).

Jobindex is not a complete library of all job posting between 2007 and 2017. Conclusions made from Jobindex data are then limited in the sense that data does not provide a complete picture of labor market demand, thus ruling out general remarks. It is possible that firms located in outer municipalitites more often than other firms choose not to post job offers in Jobindex. This would be a source of bias in any geografic related conclusion regarding demand of labor.
