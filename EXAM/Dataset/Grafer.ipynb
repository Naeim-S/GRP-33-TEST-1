```python
#Exclude "danmark" and sort by n_area and year (for nicer plot):
jobindex_area = jobindex.query(' area != "danmark"').sort_values(by=["n_area_pc","year"], ascending=False)

sns.set_palette("tab10")
sns.relplot(x="year", y="n_area_pc", 
            hue="area", 
            kind="line", 
            height=7, 
            aspect=1.5, 
            legend = "brief",
            data=jobindex_area)
plt.ylabel("Job postings per WAP")
#plt.yticks(range(0,2,20000))
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("")
```

```python
jobindex_area = jobindex.query(' area != "danmark"').sort_values(by=["n_year_pc","year"], ascending=False)

sns.set_palette("tab10")
sns.relplot(x="year", y="n_year_pc", 
          #  hue="year", 
            kind="line", 
            height=5, 
            aspect=1.4, 
            legend = "brief",
            data=jobindex_area)
plt.title("Job postings per WAP")
plt.ylabel('Job postings')
plt.xlabel('Year')
plt.ylim(0,0.14)
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("")
```
