```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns 

# Anvendte datas√¶t: unemploymentn, n_sektor11, Jobindex3_grafer
```

```python
# Figure 1

path = "https://github.com/Naeim-S/GRP33/blob/master/EXAM/Dataset/unemploymentn.xlsx?raw=true"
unemploymentn = pd.read_excel(path)

fig, ax1 = plt.subplots()
ax1.plot(unemploymentn['Unnamed: 0.1'], unemploymentn['n_year_pc'])
ax1.tick_params(axis='y')
ax1.set_xlabel('Year')
ax1.set_ylabel('Job postings per WAP (blue)')

ax2 = ax1.twinx()
ax2.plot(unemploymentn['Unnamed: 0.1'], unemploymentn['posts/unemploy'], color='red')
ax2.tick_params(axis='y')
ax2.set_ylabel('Job postings per unemployed (red)')
plt.title('Job postings per year')

fig.tight_layout()

plt.show()

# plt.savefig('Figure1.png', bbox_inches = 'tight')
```

```python
# Figure 2

n_sektor11 = pd.read_csv('https://raw.githubusercontent.com/Naeim-S/GRP33/master/EXAM/Dataset/n_sektor11')

sector_order=n_sektor11.sektor.unique()

sns.set_palette("tab10")
sns.relplot(x="year", y="0", 
            hue="sektor", 
            kind="line", 
            height=5, 
            aspect=1.4, 
            legend = "brief",
            hue_order=sector_order,
            data=n_sektor11)
plt.title("Job postings by job category")
plt.ylabel('Job postings')
plt.xlabel('Year')
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))

plt.savefig('Figure2.png', bbox_inches = 'tight')
```

```python
# Figure 3

#Insert absolute filepath from GitHub account
path2 = "https://github.com/Naeim-S/GRP33/blob/master/EXAM/Dataset/Jobindex3_grafer.xlsx?raw=true" 

cols = list(range(1,9))
jobindex = pd.read_excel(path2)

jobindex.area.unique()
area_order_=jobindex.area.unique()
sektor_order=jobindex.sector.unique()

jobindex_area = jobindex.query(' area != "danmark"').sort_values(by=["n_area_pc","year"], ascending=False)

sns.set_palette("tab10")
sns.relplot(x="year", y="n_area_pc", 
            hue="area", 
            kind="line", 
            height=4, 
            aspect=1.4, 
            legend = "brief",
            hue_order=area_order_,
            data=jobindex_area)
plt.ylabel("Job postings per WAP")
plt.ylim(0,0.175)
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")
plt.title('Job postings by region')

# plt.savefig('Figure3.png', bbox_inches = 'tight')
```

```python
# Figure 4

#Capital Area
sns.relplot(x="year", y="n_sector_area_pc", 
        hue="sector", 
        kind="line", 
        legend = "brief",        
        height=4.5, 
        aspect=1.4, 
      #  ax=ax1,
        hue_order=sektor_order,
        data=jobindex.query('area == "Region Capital Area" ').sort_values(by=["year", "n_sector_area_pc"], ascending=False))
plt.title("Job postings by sector in Region Capital Area")
plt.ylabel("Job postings per WAP")
plt.ylim(0,0.04)
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")

plt.savefig('Figure41.png', bbox_inches = 'tight')

#Zealand
sns.relplot(x="year", y="n_sector_area_pc", 
        hue="sector", 
        kind="line", 
        legend = "brief",        
        height=4.5, 
        aspect=1.4, 
      #  ax=ax2,
        hue_order=sektor_order,
        data=jobindex.query('area == "Region Zealand" ').sort_values(by=["year", "n_sector_area_pc"], ascending=False))
plt.title("Job postings by sector in Region Southern Denmark")
plt.ylabel("")
plt.ylim(0,0.04)
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")

plt.savefig('Figure42.png', bbox_inches = 'tight')

#plt.show()
```

```python
# Figure 5

#Capital Area IT 
sns.relplot(x="year", y="n_jobtype_area_pc", 
        hue="jobtype", 
        kind="line", 
        legend = "brief",        
        height=3.5, 
        aspect=1.2, 
        data=jobindex.query('area == "Region Capital Area" & sector == "IT"').sort_values(by=["year", "n_jobtype_area_pc"], ascending=False))
plt.title("a) Job postings by job type in IT in Region Capital Area")
plt.ylabel("Job postings in IT per WAP")
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")

#Zealand industry
sns.relplot(x="year", y="n_jobtype_area_pc", 
        hue="jobtype", 
        kind="line", 
        legend = "brief",        
        height=3.5, 
        aspect=1.2, 
        data=jobindex.query('area == "Region Zealand" & sector == "Industry"').sort_values(by=["year", "n_jobtype_area_pc"], ascending=False))
plt.title("c) Job postings by job type in industry in Region Zealand")
plt.ylabel("Job postings in industry per WAP")
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")

#Capital Area office work
sns.relplot(x="year", y="n_jobtype_area_pc", 
        hue="jobtype", 
        kind="line", 
        legend = "brief",        
        height=3.5, 
        aspect=1.2, 
        data=jobindex.query('area == "Region Capital Area" & sector == "Office work"').sort_values(by=["year", "n_jobtype_area_pc"], ascending=False))
plt.title("b) Job postings by job type in office work in Region Capital Area")
plt.ylabel("Job postings in office work per WAP")
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")


#Zealand trade
sns.relplot(x="year", y="n_jobtype_area_pc", 
        hue="jobtype", 
        kind="line", 
        legend = "brief",        
        height=3.5, 
        aspect=1.2, 
        data=jobindex.query('area == "Region Zealand" & sector == "Trade"').sort_values(by=["year", "n_jobtype_area_pc"], ascending=False))
plt.title("d) Job postings by job type in trade in Region Zealand")
plt.ylabel("Job postings in trade per WAP")
plt.xlim(2007,2017)
plt.xticks(range(2007,2018,2))
plt.xlabel("Year")

```
