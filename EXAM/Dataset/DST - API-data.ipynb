The data that we chose to include covers 1) development of the Gross National Product (BNP) per citizen, 2) development in unemployment by area of residence, and 3) development in working age population (WAP). Data for BNP and WAP* covers the full yearspan of 2007 - 2017 with both years included, whilst the data avilable for unemployment only covers the years 2008 to 2016. Data hails respectively from the NAN1 (Yearly National Accounting), the RAS201 (Registerbased Workforce Statistics), and the FOLK1A (Census) statistics. 

Data covering BNP and unemployment has been gathered using the API offered by DST. The source is openly availabe and yields offiicial statistical data by which usage is permitted whilst DST as the source is properly credited. As gathering the data for WAP proved overly troublesome using the API, this was simply gathered from Statistic Denmark's statistical bank.

The supplementary data is included on merit of performing procedures of comparison, sanity checking, and estimating point elasticites for the JobIndex archive-data.

________
_*FOLK1A holds information on census by each quarter of the year. Data is however limited to the earliest count being the first quarter of 2008. As thus, we have counted working age population by the 4th quarter for every year, holding 2008 Q1 as an adequate replacement for data in 2007 Q4._

```python
import requests
import numpy as np
import pandas as pd

def get_data(table_id,variables):
    base = 'https://api.statbank.dk/v1/data/{id}/JSONSTAT?lang=en'.format(id = table_id)
    
    for var in variables:
        base += '&{v}'.format(v = var) 

    response=requests.get(base)
    data_json=response.json()
    return data_json
```

```python
def df_to_markdown(*dfs, sep_line='\n---\n', **kwargs):
    """Convert pandas dataframe to markdown table."""
    import tabulate

    disable_numparse = kwargs.pop('disable_numparse', True)
    tablefmt = kwargs.pop('tablefmt', 'pipe')
    headers = kwargs.pop('headers', 'keys')
    
    for df in dfs:
        print(tabulate.tabulate(df, tablefmt=tablefmt, headers=headers,
                                disable_numparse=disable_numparse, **kwargs))
        if sep_line is not None:
            print(sep_line)
            
            
df_to_markdown(FOLK1A)
```
