```python
# Scraping

sektor = {"it":["database", "itdrift", "itkurser", "itledelse",
                "internet", "systemudvikling", "telekom", "virksomhedssystemer"],
         "ingenioer":["byggeteknik", "elektronik", "kemi", "teknikledelse",  
                     "maskiningenioer", "medicinal", "produktionsteknik"],
         "ledelse":["detailledelse", "freelancekonsulent", "hrkurser", "itledelse",
                   "institutions", "leder", "teknikledelse", "personale", "projektledelse",
                   "salgschef", "topledelse", "virksomhedsudvikling", "oekonomichef"],
         "handel": ["bud", "boernepasning", "detailhandel", "ejendomsservice",
                    "frisoer", "hotel", "rengoering", "service", "sikkerhed"],
         "industri": ["blik", "byggeri", "elektriker", "industri", "jern", "lager", "landbrug",
                     "maling", "mekanik", "naeringsmiddel", "tekstil", "transport", "traeindustri",
                     "toemrer"],
         "salg": ["design", "ejendomsmaegler", "grafisk", "kommunikation", "kultur", "marketing",
                  "salg", "salgskurser", "franchise", "telemarketing"],
         "undervisning": ["bibliotek", "forskning", "laerer", "paedagog", "voksenuddannelser"],
         "kontor": ["akademisk", "ejendomsmaegler", "ejendomsservice", "finans", "indkoeb",
                   "jura", "kontor", "kontorkurser", "kontorelev", "logistik", "offentlig",
                   "oversaettelse", "sekretaer", "oekonomi"],
         "social": ["laege", "laegesekretaer", "pleje", "psykologi", "socialraadgivning",
                    "sygeplejske", "tandlaege", "teknisksundhed", "terapi"],
         "oevrige": ["elev", "forsvar", "frivilligt", "student", "studiepraktik", "oevrige", "kurseroevrige"]}
område = ["storkoebenhavn", "nordsjaelland", "region-sjaelland", \
          "fyn", "region-nordjylland", "region-midtjylland", \
         "sydjylland", "bornholm", "skaane", "groenland", "faeroeerne" \
         "udlandet", "danmark"]

# Generér funktioner
    # n_jobtype_område
def NJO(sektor, jobtype, område, år):
    url1 = 'https://www.jobindex.dk/jobsoegning/'
    url2 = '/?jobage=archive&maxdate='
    url3 = '1231&mindate='
    url4 = '0101'
    return url1 + str(sektor) + '/' + str(jobtype) + '/' + str(område) + url2 + str(år) + url3 + str(år) + url4

    # n_sektor_område
def NSO(sektor, område, år):
    url1 = 'https://www.jobindex.dk/jobsoegning/'
    url2 = '/?jobage=archive&maxdate='
    url3 = '1231&mindate='
    url4 = '0101'
    return url1 + str(sektor) + '/' + str(område) + url2 + str(år) + url3 + str(år) + url4

    # n_sektor
def NS(sektor, år):
    url1 = 'https://www.jobindex.dk/jobsoegning/'
    url2 = '?jobage=archive&maxdate='
    url3 = '1231&mindate='
    url4 = '0101'
    return url1 + str(sektor) + url2 + str(år) + url3 + str(år) + url4

    # n_område
def NO_(område, år):
    url1 = 'https://www.jobindex.dk/jobsoegning/'
    url2 = '?jobage=archive&maxdate='
    url3 = '1231&mindate='
    url4 = '0101'
    return url1 + str(område) + url2 + str(år) + url3 + str(år) + url4

    # n_år
def NÅ(år):
    url1 = 'https://www.jobindex.dk/jobsoegning?maxdate='
    url2 = '1231&mindate='
    url3 = '0101&jobage=archive'
    return url1 + str(år) + url2 + str(år) + url3
```

```python
# Scraping

år = [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]

n_jobtype_område_ = []
n_sektor_område_ = []
n_område_ = []
n_år_ = []

jobt_ = []
sektor_ = []
område_ = []
år_ = []


# Running loop:

for z in år:
# n_år:
    urlNÅ = NÅ(z)

    responseNÅ = requests.get(urlNÅ)
    htmlNÅ = responseNÅ.text
    jobindexNÅ = BeautifulSoup(htmlNÅ,'lxml')


    tal1NÅ = str(jobindexNÅ.find('h2'))
    if """<h2 class="search-title">""" in tal1NÅ:
        tal2NÅ = re.sub("""<h2 class="search-title">""", '', tal1NÅ)
        tal3NÅ = re.sub(""" job matcher din søgning</h2>""", '', tal2NÅ)
        tal4NÅ = int(tal3NÅ.replace('.',''))

        for x in område:
# n_område:
            urlNO = NO_(x, z)

            responseNO = requests.get(urlNO)
            htmlNO = responseNO.text
            jobindexNO = BeautifulSoup(htmlNO,'lxml')


            tal1NO = str(jobindexNO.find('h2'))
            if """<h2 class="search-title">""" in tal1NO:
                tal2NO = re.sub("""<h2 class="search-title">""", '', tal1NO)
                tal3NO = re.sub(""" job matcher din søgning</h2>""", '', tal2NO)
                tal4NO = int(tal3NO.replace('.',''))

# n_sektor_område:
                for key in sektor:
                    urlNSO = NSO(key, x, z)

                    responseNSO = requests.get(urlNSO)
                    htmlNSO = responseNSO.text
                    jobindexNSO = BeautifulSoup(htmlNSO,'lxml')


                    tal1NSO = str(jobindexNSO.find('h2'))
                    if """<h2 class="search-title">""" in tal1NSO:
                        tal2NSO = re.sub("""<h2 class="search-title">""", '', tal1NSO)
                        tal3NSO = re.sub(""" job matcher din søgning</h2>""", '', tal2NSO)
                        tal4NSO = int(tal3NSO.replace('.',''))

# n_jobtype_område:
                        for value in sektor[key]:
                            urlNJO = NJO(key, value, x, z)

                            responseNJO = requests.get(urlNJO)         # Opens link
                            htmlNJO = responseNJO.text
                            jobindexNJO = BeautifulSoup(htmlNJO,'lxml')


                            tal1NJO = str(jobindexNJO.find('h2'))                                          # Searching for tag 'h2' to locate wanted number
                            if """<h2 class="search-title">""" in tal1NJO:                                 # Checking if correct string is located
                                tal2NJO = re.sub("""<h2 class="search-title">""", '', tal1NJO)             # Removing first part of string
                                tal3NJO = re.sub(""" job matcher din søgning</h2>""", '', tal2NJO)         # Removing second part of string
                                tal4NJO = int(tal3NJO.replace('.',''))                                     # Removing dot and converting to integer


# Appending values to lists:    
                                n_jobtype_område_.append(tal4NJO)
                                n_sektor_område_.append(tal4NSO)
                                n_område_.append(tal4NO)
                                n_år_.append(tal4NÅ)


                                jobt_.append(value)
                                sektor_.append(key)
                                område_.append(x)
                                år_.append(z)

                                time.sleep(0.3)

# Appending lists to dataframe:
container1417 = pd.DataFrame(
    {'n_jobtype_area_': n_jobtype_område_,
     'n_year': n_år_,
     'n_sektor_area': n_sektor_område_,
     'n_area': n_område_,
     'jobtype': jobt_,
     'sektor': sektor_,
     'area': område_,
     'year': år_
    })

container1417.to_csv('con1417.csv')
```

```python
import requests
import numpy as np
import pandas as pd

def get_data(table_id,variables):
    base = 'https://api.statbank.dk/v1/data/{id}/JSONSTAT?lang=en'.format(id = table_id)
    
    for var in variables:
        base += '&{v}'.format(v = var) 

    response=requests.get(base)
    data_json=response.json()
    return data_json
```

## Dataframe to markdown

```python
def df_to_markdown(*dfs, sep_line='\n---\n', **kwargs):
    """Convert pandas dataframe to markdown table."""
    import tabulate

    disable_numparse = kwargs.pop('disable_numparse', True)
    tablefmt = kwargs.pop('tablefmt', 'pipe')
    headers = kwargs.pop('headers', 'keys')
    
    for df in dfs:
        print(tabulate.tabulate(df, tablefmt=tablefmt, headers=headers,
                                disable_numparse=disable_numparse, **kwargs))
        if sep_line is not None:
            print(sep_line)
            
            
df_to_markdown(FOLK1A)
```
